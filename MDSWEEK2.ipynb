```python
import numpy as np
```

### NumPy Array Indexing

```python
# Create a 1D array for indexing demonstrations
arr_1d = np.array([102, 2, 3, 4, 5, 6, 7, 8, 9])
print("Original 1D array:", arr_1d)
```

```python
# Accessing a single element by positive index (first element)
print("Element at index 0:", arr_1d[0])
```

```python
# Accessing a single element by negative index (last element)
print("Element at last index:", arr_1d[-1])
```

```python
# Create a 2D array for indexing demonstrations
arr_2d = np.array([[102, 20, 30], [40, 50, 60], [70, 80, 90]])
print("\nOriginal 2D array:\n", arr_2d)
```

```python
# Accessing an element in a 2D array (row 0, column 1)
print("Element at [0, 1]:", arr_2d[0, 1])
```

```python
# Accessing another element in a 2D array (row 2, column 2)
print("Element at [2, 2]:", arr_2d[2, 2])
```

```python
# Boolean indexing: select elements greater than 50
bool_indexed_arr = arr_2d[arr_2d > 50]
print("\nElements greater than 50 (Boolean indexing):", bool_indexed_arr)
```

### NumPy Array Slicing

```python
# Create a 1D array for slicing demonstrations
arr_1d_slice = np.array([102, 1, 2, 3, 4, 5, 6, 7, 8, 9])
print("Original 1D array:", arr_1d_slice)
```

```python
# Basic slicing: elements from index 2 up to (but not including) index 5
print("Slice arr_1d_slice[2:5]:", arr_1d_slice[2:5])
```

```python
# Slice from the beginning up to (but not including) index 4
print("Slice arr_1d_slice[:4]:", arr_1d_slice[:4])
```

```python
# Slice from index 6 to the end
print("Slice arr_1d_slice[6:]:", arr_1d_slice[6:])
```

```python
# Slice with a step: every other element
print("Slice with step (arr_1d_slice[::2]):", arr_1d_slice[::2])
```

```python
# Reverse the array using slicing
print("Reversed array (arr_1d_slice[::-1]):", arr_1d_slice[::-1])
```

```python
# Create a 2D array for slicing demonstrations
arr_2d_slice = np.array([[102, 20, 30, 40], [50, 60, 70, 80], [90, 100, 110, 120]])
print("\nOriginal 2D array:\n", arr_2d_slice)
```

```python
# Slicing 2D: Get rows 0 and 1, all columns
print("Rows 0-1, all columns (arr_2d_slice[0:2, :]):\n", arr_2d_slice[0:2, :])
```

```python
# Slicing 2D: Get all rows, columns 1 and 2
print("All rows, columns 1-2 (arr_2d_slice[:, 1:3]):\n", arr_2d_slice[:, 1:3])
```

```python
# Slicing 2D: Get a sub-matrix (rows 1-2, cols 0-1)
print("Sub-matrix (rows 1-2, cols 0-1):\n", arr_2d_slice[1:3, 0:2])
```

### NumPy Array Reshaping

```python
# Create a 1D array for reshaping demonstrations
arr_reshape = np.arange(12) # Creates array [0, 1, ..., 11]
arr_reshape[0] = 102 # Set the first element to 102
print("Original 1D array:", arr_reshape)
```

```python
# Reshape the 1D array into a 3x4 2D array
arr_reshaped_3x4 = arr_reshape.reshape(3, 4)
print("\nReshaped to 3x4:\n", arr_reshaped_3x4)
```

```python
# Reshape the 1D array into a 4x3 2D array
arr_reshaped_4x3 = arr_reshape.reshape(4, 3)
print("\nReshaped to 4x3:\n", arr_reshaped_4x3)
```

```python
# Reshape with an inferred dimension (-1): 2 rows, columns inferred
arr_inferred_cols = arr_reshape.reshape(2, -1)
print("\nReshaped with inferred columns (2 rows):\n", arr_inferred_cols)
```

```python
# Reshape with an inferred dimension (-1): 3 columns, rows inferred
arr_inferred_rows = arr_reshape.reshape(-1, 3)
print("\nReshaped with inferred rows (3 columns):\n", arr_inferred_rows)
```

```python
# Flatten an array using the .flatten() method (returns a copy)
print("\nFlattened array (using .flatten()):", arr_reshaped_3x4.flatten())
```

```python
# Flatten an array using the .ravel() method (returns a view if possible)
print("\nRaveled array (using .ravel()):", arr_reshaped_3x4.ravel())
```

### NumPy Array Joining

```python
# Create two 1D arrays for joining demonstrations
arr_join_1d_1 = np.array([102, 2, 3])
arr_join_1d_2 = np.array([4, 5, 6])
print("1D Array 1:", arr_join_1d_1)
print("1D Array 2:", arr_join_1d_2)
```

```python
# Concatenate 1D arrays along axis 0 (default behavior)
joined_1d_arr = np.concatenate((arr_join_1d_1, arr_join_1d_2))
print("Concatenated 1D arrays:", joined_1d_arr)
```

```python
# Create two 2D arrays for joining demonstrations
arr_join_2d_1 = np.array([[102, 20], [30, 40]])
arr_join_2d_2 = np.array([[50, 60], [70, 80]])
print("\n2D Array 1:\n", arr_join_2d_1)
print("2D Array 2:\n", arr_join_2d_2)
```

```python
# Concatenate 2D arrays along rows (axis=0)
joined_2d_rows = np.concatenate((arr_join_2d_1, arr_join_2d_2), axis=0)
print("\nJoined along rows (axis=0):\n", joined_2d_rows)
```

```python
# Concatenate 2D arrays along columns (axis=1)
joined_2d_cols = np.concatenate((arr_join_2d_1, arr_join_2d_2), axis=1)
print("\nJoined along columns (axis=1):\n", joined_2d_cols)
```

```python
# Vertically stack arrays using np.vstack()
stacked_v = np.vstack((arr_join_2d_1, arr_join_2d_2))
print("\nVertically stacked (vstack):\n", stacked_v)
```

```python
# Horizontally stack arrays using np.hstack()
stacked_h = np.hstack((arr_join_2d_1, arr_join_2d_2))
print("\nHorizontally stacked (hstack):\n", stacked_h)
```

```python
# Depth stack (dstack) for creating 3D arrays from 2D arrays
stacked_d = np.dstack((arr_join_2d_1, arr_join_2d_2))
print("\nDepth stacked (dstack):\n", stacked_d)
print("Shape of dstacked array:", stacked_d.shape)
```

### NumPy Array Splitting

```python
# Create a 1D array for splitting demonstrations
arr_split_1d = np.arange(10)
arr_split_1d[0] = 102 # Set the first element to 102
print("Original 1D array:", arr_split_1d)
```

```python
# Split the 1D array into 2 equal parts
split_1d_part1, split_1d_part2 = np.split(arr_split_1d, 2)
print("\nSplit into 2 equal parts:", split_1d_part1, split_1d_part2)
```

```python
# Split the 1D array at specific indices (before index 2 and before index 7)
split_1d_at_indices = np.split(arr_split_1d, [2, 7])
print("\nSplit at indices [2, 7]:", split_1d_at_indices)
```

```python
# Create a 2D array for splitting demonstrations
arr_split_2d = np.arange(16).reshape(4, 4)
arr_split_2d[0,0] = 102 # Set the first element to 102
print("\nOriginal 2D array:\n", arr_split_2d)
```

```python
# Vertically split (vsplit) the 2D array into 2 equal rows
vs_2d_part1, vs_2d_part2 = np.vsplit(arr_split_2d, 2)
print("\nVertical split (vsplit) into 2 parts:\n", vs_2d_part1, "\n", vs_2d_part2)
```

```python
# Horizontally split (hsplit) the 2D array into 2 equal columns
hs_2d_part1, hs_2d_part2 = np.hsplit(arr_split_2d, 2)
print("\nHorizontal split (hsplit) into 2 parts:\n", hs_2d_part1, "\n", hs_2d_part2)
```

```python
# Unequal split using array_split (can handle non-divisible splits) along rows (axis=0)
split_unequal_rows = np.array_split(arr_split_2d, 3, axis=0)
print("\nUnequal split into 3 rows (array_split):\n")
for i, part in enumerate(split_unequal_rows):
    print(f"Part {i+1}:\n{part}")
    print("---")
```

### Universal Functions (UFuncs)

```python
# Create a 1D array for UFunc demonstrations
arr_ufunc = np.array([102, 2, 3, 4, 5])
print("Original array for UFuncs:", arr_ufunc)
```

```python
# Element-wise addition with a scalar
print("Addition (arr_ufunc + 5):", arr_ufunc + 5)
```

```python
# Element-wise multiplication with a scalar
print("Multiplication (arr_ufunc * 2):", arr_ufunc * 2)
```

```python
# Element-wise square root using np.sqrt()
# Note: np.sqrt(102) is ~10.099
print("Square root (np.sqrt(arr_ufunc)):", np.sqrt(arr_ufunc))
```

```python
# Element-wise exponentiation (e^x) using np.exp()
print("Exponentiation (np.exp(arr_ufunc)):", np.exp(arr_ufunc))
```

```python
# Element-wise sine using np.sin()
print("Sine (np.sin(arr_ufunc)):", np.sin(arr_ufunc))
```

```python
# Element-wise cosine using np.cos()
print("Cosine (np.cos(arr_ufunc)):", np.cos(arr_ufunc))
```

```python
# Element-wise natural logarithm using np.log()
print("Logarithm (np.log(arr_ufunc)):", np.log(arr_ufunc))
```

```python
# Create a float array for rounding UFunc demonstrations
arr_float_ufunc = np.array([102.0, 2.3, 3.8, 4.1])
print("\nFloat array for rounding:", arr_float_ufunc)
```

```python
# Element-wise floor using np.floor()
print("Floor (np.floor(arr_float_ufunc)):", np.floor(arr_float_ufunc))
```

```python
# Element-wise ceiling using np.ceil()
print("Ceil (np.ceil(arr_float_ufunc)):", np.ceil(arr_float_ufunc))
```

```python
# Element-wise rounding using np.round()
print("Round (np.round(arr_float_ufunc)):", np.round(arr_float_ufunc))
```

```python
# Create two arrays for binary UFunc demonstrations
arr_ufunc_a = np.array([102, 2, 3])
arr_ufunc_b = np.array([4, 5, 6])
print("\nArray A for binary UFuncs:", arr_ufunc_a)
print("Array B for binary UFuncs:", arr_ufunc_b)
```

```python
# Element-wise addition of two arrays
print("Element-wise addition (arr_ufunc_a + arr_ufunc_b):", arr_ufunc_a + arr_ufunc_b)
```

```python
# Element-wise maximum of two arrays using np.maximum()
print("Element-wise maximum (np.maximum(arr_ufunc_a, arr_ufunc_b)):", np.maximum(arr_ufunc_a, arr_ufunc_b))
```

### Mathematical Methods

```python
# Create a 1D array for mathematical method demonstrations
arr_math = np.array([102, 2, 3, 4, 5])
print("Original array for mathematical methods:", arr_math)
```

```python
# Minimum element in the array using .min()
print("Minimum element (arr_math.min()):", arr_math.min())
```

```python
# Maximum element in the array using .max()
print("Maximum element (arr_math.max()):", arr_math.max())
```

```python
# Mean (average) of elements in the array using .mean()
print("Mean of elements (arr_math.mean()):", arr_math.mean())
```

```python
# Standard deviation of elements in the array using .std()
print("Standard deviation (arr_math.std()):", arr_math.std())
```

```python
# Product of all elements in the array using .prod()
print("Product of elements (arr_math.prod()):", arr_math.prod())
```
